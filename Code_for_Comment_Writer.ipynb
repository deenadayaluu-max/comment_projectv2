```python
import csv

course_name = 'Python II'

course_description = 'This course builds on the skills and concepts learned in our previous computer science courses.  The topics challenge students to explore how computing and technology can impact the world, with a unique focus on creative problem solving and real-world applications.  A successful student will know more advanced data structures and object-oriented programming basics.  A successful student will be able to communicate with popular APIs including Twitter and Google in order to “grab” data and to organize the data and display it.'


def reading_file(file_name):
    """
    Written by Taylor
    Takes in a given file name and creates a
    list of the informaton listed in the csv file
    """
    things = []
    with open(file_name) as file:
        header = next(file)
        data = csv.reader(file, delimiter=',')
        for info in data:
            things.append(info)
    return things


DATA = reading_file('all_info2.csv')


def adjectives(student_info):
    """
    Written by Evie and Taylor
    Takes in a student's name and returns a list of
    the adjectives associated with that student.
    """
    adj = student_info[2].split(', ')
    adjectives = []
    for word in adj:
        adjectives.append(word)
    return adjectives


def learning_outcomes():
    """
    Written by everyone
    Reads in the learning outcomes and
    returns a dictionary of the outcome number and the outcome
    """
    outcomes_dict = {}
    with open('pythonlearningoutcomes.csv') as file:
        header = next(file)
        outcomes = csv.reader(file, delimiter=',')
        counter = 1
        for outcome in outcomes:
            if counter not in outcomes_dict:
                outcomes_dict[counter] = outcome[0]
            counter += 1
    return outcomes_dict


OUTCOMES = learning_outcomes()


def strength_1(student_info):
    """
    Written by Uma
    Organizes feedback according to information
    gathered from the data about a student's strengths.
    """
    outcome_num = int(student_info[3])
    outcome_description = OUTCOMES[outcome_num]
    comment = student_info[4]
    return f"Strength:\nLearning Outcome {outcome_num} - {outcome_description}\nComment: {comment}"


def strength_2(student_info):
    """
    Written by Evie
    Organizes feedback according to information
    gathered from the data about a student's strengths.
    """
    outcome_num = int(student_info[5])
    outcome_description = OUTCOMES[outcome_num]
    comment = student_info[6]
    return f"\n\nStrength:\nLearning Outcome {outcome_num} - {outcome_description}\nComment: {comment}"


def area_of_growth(student_info):
    """
    Written by Taylor
    Organizes feedback according to information
    gathered from the data about a student's areas of growth.
    """
    outcome_num = int(student_info[7])
    outcome_description = OUTCOMES[outcome_num]
    comment = student_info[8]
    return f"\n\nArea of Growth:\nLearning Outcome {outcome_num} - {outcome_description}\nComment: {comment}"


def closing_section(student_info):
    """
    Written by everyone
    Based on the student's info, returns a closing section
    about the student's overall performance in the class.
    """
    ADJECTIVES = adjectives(student_info)
    sentence = student_info[-1]
    if len(ADJECTIVES) >= 2:
        word1 = ADJECTIVES[0]
        word2 = ADJECTIVES[1]
        return f'\n\nClosing: {student_info[0].title()}! Some words I would use to describe you as a student are {word1} and {word2}. I look forward to seeing you next semester. {sentence}. Have a great break!'
    else:
        word = ADJECTIVES[0]
        return f'\n\nClosing: {student_info[0].title()}! One word I would use to describe you as a student is {word}. I look forward to seeing you next semester. {sentence} Enjoy the break!'


def overall_grade(student_info):
    """
    Written by Taylor
    Returns the overall grade of the student
    """
    grade = student_info[1]
    return grade


def final_comment(student_info):
    """
    Written by everyone
    Using all the student info provided, returns a final comment
    that takes into account strengths, areas of growth, and a closing section.
    """
    with open(f"{student_info[0]}.txt", 'w') as comment_file:
        comment_file.write(
            f"Course Name: {course_name}\n{course_description}\n\n")
        comment_file.write(strength_1(student_info))
        comment_file.write(strength_2(student_info))
        comment_file.write(area_of_growth(student_info))
        comment_file.write(closing_section(student_info))
        comment_file.write(f"\n\nOverall grade: {overall_grade(student_info)}")


# Evie
for student_info in DATA:
    final_comment(student_info)

# Pep 8 standards: Uma
```
